```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([10,20,30])
#x=np.random.normal(1,10,3)
plt.pie(x, labels=['boy','girls','students'],labeldistance=0.5,explode=[0.2,0,0])
plt.legend(title="")#legend will display all the lables in colume
plt.show()
```
