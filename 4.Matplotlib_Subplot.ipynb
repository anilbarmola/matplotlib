```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,5,2,6,3])
y=np.array([1,5,2,6,3])
plt.subplot(1,2,1)#1 row and 2 column and last one is subplot number
plt.plot(x,y)
plt.title("Title1", loc='left')
#SUBPLOT
x=np.array([10,50,20,60,30])
y=np.array([1,5,2,6,3])
plt.subplot(1,2,2)
plt.grid()#show grid
plt.plot(x,y ,marker='o', mfc='r')
plt.title("Title2")
plt.show()
```
