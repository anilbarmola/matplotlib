```python
import matplotlib.pyplot as plt 
import numpy as np  
ypoint=np.array([10,20,30,40,50])
plt.plot(ypoint,marker ='o', ms=20 , mec="r", mfc="g")
#ms=marker size and mec=marker border color and mfc=marker body color
plt.show()
```

```python
y=[1,5,6,9]
x=[1,2,3,4]
plt.plot(y)
plt.plot(x)
plt.grid()
plt.show()
```

```python
#Multiple Lines
import matplotlib.pyplot as plt
import numpy as np
y1point=np.array([10,20,30,25,16])
y2point=np.array([12,15,26,18,20,12])
plt.plot(y1point, marker='o')
plt.plot(y2point, mfc='b', color='r',marker='*')
plt.show()
```
